# Proyecto: Grafo de los persponajes de The Last of Us

Mi proyecto se centrara en realizar un grafos sobre las relaciones de los personajes de la saga de The Last of Us incluyendo las partes I y II además del DLC Left Behind, se realizaran grafos individuales de cada juego de la saga y finalmente un grafo que incluya todos los personajes de los grafos anteriores. Luego dibujaremos el grafo y veremos los conjuntos de nodos y vertices ordenados que tenemos dentro de nuestros grafos, vecinos que están conectados al nodo con más conexiones, los grados de todos los nodos y la excentricidad de estos mismos.

## Conclusión

Un grafo es una estructura matemática que consiste en un conjunto de nodos (vértices) conectados por arcos (bordes). En el contexto de conectar personajes dentro de una historia, un grafo puede ser utilizado para representar las relaciones y conexiones entre ellos de manera visual y estructurada.

Un grafo puede representar visualmente cómo están conectados los personajes entre sí. Cada personaje puede ser un nodo, y las conexiones entre ellos (amistad, enemistad, parentesco, etc.) pueden ser representadas por arcos entre los nodos.

Utilizando técnicas de análisis de redes sociales, un grafo puede revelar quiénes son los personajes más centrales (más conectados) y quiénes están más aislados en la trama. Esto puede ayudar a los escritores a enfocarse en las relaciones más importantes o subrayar la importancia de ciertos personajes en la historia.

Al mapear las conexiones entre personajes, los escritores pueden visualizar cómo se entrelazan diferentes tramas y subtramas. Esto facilita la planificación y la gestión de la complejidad narrativa al asegurarse de que cada relación y cada personaje tenga un propósito claro en la historia.

En resumen, un grafo es una herramienta poderosa para visualizar y gestionar las complejas relaciones entre personajes dentro de una historia. Ayuda a los escritores a estructurar y desarrollar narrativas de manera más efectiva, asegurando que cada personaje y cada conexión contribuyan de manera significativa al desarrollo general de la trama.
